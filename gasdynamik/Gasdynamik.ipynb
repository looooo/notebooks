## Lösung der beschreibenden Differentialgleichung

Die Ausgangslage stellt die Differentialgleichung in der folgenden Form dar: [1] S.136 6.41
$$
\tag{1}
\int\limits_{\vartheta^*}\limits^{\vartheta} d \vartheta = 
\int\limits_{1}\limits^{M^*} \frac{\sqrt{M^2-1}}{1+c \cdot M^2} \frac{dM}{M}
\ \ \text{mit} \ \ c = \frac{\kappa-1}{2}
$$

Substietution von $a = \sqrt{M^2-1}$ liefert:
$$
\vartheta - \vartheta^* = \int\limits_{1}\limits^{\sqrt{M^2-1}} \frac{a}{1+c \cdot (a^2 + 1)} \frac{a}{a^2 + 1} da = 
\int\limits_{0}\limits^{\sqrt{M^2-1}} \frac{a^2}{(a^2+1)(a^2 \cdot c + c + 1)} da
$$

In Partialbrüche zerlegt:
$$
\vartheta - \vartheta^* = \int\limits_{0}\limits^{\sqrt{M^2-1}} \left( \frac{-1}{a^2+1}  + \frac{c+1}{a^2 \cdot c + c + 1} \right) da
$$

Das erste Integral kann direkt ausgewertet werden:
$$
\int\limits_{0}\limits^{\sqrt{M^2-1}} \frac{-1}{a^2+1} da = -\arctan \left( \sqrt{M^2-1} \right)
$$

Für das zweite Integral führt man die Substittion $b^2 = \frac{a^2 \cdot c}{c + 1}$ ein, was zu folgender Formulierung führt:
$$
\int\limits_{0}\limits^{\sqrt{M^2-1}} \frac{c+1}{a^2 \cdot c + c + 1} da =
\int\limits_{0}\limits^{\sqrt{M^2-1}} \frac{1} {\frac{a^2 \cdot c}{c+1} + 1} da =
\sqrt{\frac{c+1}{c}} \int\limits_{0}\limits^{\sqrt{M^2-1} \cdot \sqrt{\frac{c}{c + 1}} } \frac{1} {b^2 + 1} db = \\
=\sqrt{\frac{\kappa+1}{\kappa-1}} \arctan \sqrt{\frac{\kappa-1}{\kappa+1} \left( M^2-1 \right)}
$$

Daraus ergibt sich [1] S.136 6.42
$$
\tag{2}
\vartheta - \vartheta^* = \pm\left[
\sqrt{\frac{\kappa+1}{\kappa-1}} \arctan \sqrt{\frac{\kappa-1}{\kappa+1} \left( M^2-1 \right)} -
\arctan \left( \sqrt{M^2-1} \right) \right]
$$

## Prandtl Meyer Eckenumströmung
Die Strömung wird ausgehend von einem Winkel $\vartheta_0 = 0$ abgelenkt. Vom Eck senkrecht nach oben ausgehend wird der Winkel $\lambda$ gemessen. $\vartheta$ entspricht dem Winkel zwischen Tangente an die Stromlinie und der Horizontalen.

<img src="ecken_umstroemung.png" alt="ecken umströmung"/>


Unter Verwendung der Beziehung $\ \ \tan{\alpha} = \frac{1}{\tan{\beta}}=\frac{1}{\sqrt{M^2-1}}$ ergibt sich aus (2):
$$
\tag{3}
\vartheta = - \left[
\sqrt{\frac{\kappa+1}{\kappa-1}} \arctan{\left( \sqrt{\frac{\kappa-1}{\kappa+1}} \tan{\beta} \right)} - \beta \right]
$$

Weiters gelten folgende Beziehungen:
$$
\lambda + \alpha - \vartheta = \frac{\pi}{2} \ \ \
\beta = \frac{\pi}{2} - \alpha \\
\lambda + \beta = \vartheta
$$

Eingesetzt in (3):

$$
\tag{4}
\lambda = -\sqrt{\frac{\kappa+1}{\kappa-1}} \arctan{\left( \sqrt{\frac{\kappa-1}{\kappa+1}} \tan{\beta} \right)} \rightarrow
\tan{\left( \lambda \sqrt{\frac{\kappa-1}{\kappa+1}} \right)} = - \tan{\beta} \sqrt{\frac{\kappa-1}{\kappa+1}}
$$

Daraus folgt der Zusammenhang zwischen $\lambda$ und $\vartheta$:

$$
\tag{5}
\vartheta = \lambda -\arctan{\left[ \tan{\left( \lambda \sqrt{\frac{\kappa-1}{\kappa+1}} \right)} \sqrt{\frac{\kappa+1}{\kappa-1} } \right]}
$$

Ein weiterer Zusammenhang zwischen $\beta$, $\lambda$ und $r$ bekommt man durch die Betrachtung einer infinitesimale Winkeländerung $d\lambda$:

<img src="inf_winkel.png" alt="inf_winkel"/>

$$
\frac{dr}{\cos{\alpha}} = \frac{r \cdot d\lambda}{\sin{\alpha}} \ \ \ 
\tan{\alpha} = \frac{r \cdot d\lambda}{dr} \ \ \ 
\tan{\beta} = \frac{dr}{r \cdot d\lambda}
$$


Mit (4) erhält man somit die folgende Gleichung die jetzt integriert werden kann:

$$
\tan{\left( \lambda \sqrt{\frac{\kappa-1}{\kappa+1}} \right)} = \frac{dr}{r \cdot d\lambda} \sqrt{\frac{\kappa-1}{\kappa+1}}\\
\int\limits_{0}\limits^{\lambda}
\tan{\left( \lambda \sqrt{\frac{\kappa-1}{\kappa+1}} \right)} \cdot d\lambda = 
\int\limits_{r_0}\limits^{r}
\sqrt{\frac{\kappa-1}{\kappa+1}} \frac{dr}{r} \\
-\sqrt{\frac{\kappa+1}{\kappa-1}} ln \cos{\left( \lambda \sqrt{\frac{\kappa-1}{\kappa+1}} \right)} = 
\sqrt{\frac{\kappa-1}{\kappa+1}} ln \frac{r}{r_0} \\
-\frac{\kappa+1}{\kappa-1} ln \cos{\left( \lambda \sqrt{\frac{\kappa-1}{\kappa+1}} \right)} = ln \frac{r}{r_0}
$$

Daraus ergibt sich eine Funktion für $r$ in Abhängigkeit von $\lambda$ und $r_0$:
$$
\tag{6}
r = r_0 \cdot \cos{\left( \lambda \sqrt{\frac{\kappa-1}{\kappa+1}} \right)}^{-\frac{\kappa+1}{\kappa-1}}
$$

## Auswertung der Funktion

```python
# Module
import numpy as np
from numpy import sin, cos, tan, arctan, sqrt
import matplotlib.pyplot as plt

# Einstellungen
%matplotlib inline
plt.rcParams['figure.figsize'] = 14, 5

# Konstanten
kappa = 1.4
k1 = sqrt((kappa - 1) / (kappa + 1))

# Parameter
r_min = 0
r_max = 1
r_num = 5
lambda_end = np.pi / 2.5
num_faecher = 5


def theta(lambda_):
    """Funktion (5): Abströmwinkel theta abhängig von lambda"""
    return lambda_ - arctan(tan(lambda_ * k1) / k1)

def radius(lambda_, r0=0.1):
    """Funktion (6): Radius abhängig von lambda und dem Radius r0 bei theta = 0"""
    return r0 * (np.cos(lambda_ * k1) ** -(1 / k1**2))


# Auswertung der Funktion an den Stellen lambda_
lambda_array = np.linspace(0, lambda_end, 100)

# Auswertung der Funktion für die Anfangswerte für r0
r0_array = np.linspace(r_min, r_max, r_num)

# der Winkel zwischen den Platten (theta)
theta_end = theta(lambda_array[-1])

fig = plt.figure()
ax = fig.add_subplot(111)

for r0 in r0_array:
    r = np.array([radius(lambda_, r0) for lambda_ in lambda_array])
    
    # Umrechnen in xy Koordinatensystem
    x, y = list(r * np.sin(lambda_array)), list(r * np.cos(lambda_array))
    
    # Anhängen der Punkte für die Stromlinie vor und nach der Ablenkung
    x = [-2] + x + [x[-1] + cos(theta_end) * 2]
    y = [r0] + y + [y[-1] - sin(theta_end) * 2]
    
    # Plotten der Stromlinie
    ax.plot(x, y, c="0", linewidth=1 + float(r0==0) * 2)
    
# Plot
for lambda_ in np.linspace(0, lambda_end, num_faecher):
    r = radius(lambda_, r0_array[-1])
    ax.plot([0, r * sin(lambda_)], [0, r * cos(lambda_)], c="grey")


ax.set_aspect('equal', 'datalim')
plt.grid()
plt.show()
```
